```python
def getData():
    url = "https://www.worldometers.info/world-population/population-by-country/"
    
    from requests import get  # to make GET request

    response = get(url)
    #print(response.encoding)
        
    from bs4 import BeautifulSoup    
    soup = BeautifulSoup(response.content.decode('euc-kr'), 'html.parser')
    
    return soup
    
```

```python
dd = getData()
```

```python
dd#example2
contents = dd.select(' #example2 ') #'.type_2 > tr')
```

```python
cc = contents[0].select("tr")
ddd = []
for ii in cc:
    ddd2 = []
    jj = ii.select("td")
    for j in jj:
        ddd2.append( j.text.replace("\n","").replace("\t","").replace(",","") )
    ddd.append( ddd2 )
```

```python
dddd = {}
for i in ddd:
    try:
        dddd[i[1]] = i[2]
    except:
        continue

```

```python
dddd
```

```python
import json
from datetime import datetime
c = datetime.now().strftime("%Y%m%d%H%M")

with open("./population_"+c+".json", "w") as json_file:
    json.dump(dddd, json_file,ensure_ascii=False)
```
